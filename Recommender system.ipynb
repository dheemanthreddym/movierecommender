```python
import ast 

def convert(text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
def convert_cast(text):
    l=[]
    counter=0
    for i in ast.literal_eval(text):
        if counter<3:
            l.append(i['name'])
        counter+=1
    return l
```

```python
def fetch_director(text):
    l = []
    
    if pd.isna(text):
        return l
    
    # If already a list, don't evaluate again
    if isinstance(text, list):
        data = text
    else:
        try:
            data = ast.literal_eval(text)
        except:
            return l
    
    for i in data:
        if i.get('job') == 'Director':
            l.append(i.get('name'))
            break
            
    return l
```

```python
movies['director'] = movies['crew'].apply(fetch_director)
```

```python
import nltk
```

```python
from nltk.stem import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stems(text):
    l=[]
    for i in text.split():
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
new_df['tags']=new_df['tags'].apply(stems)
```

```python
from sklearn.feature_extraction import text
```

```python
cv = text.CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector=cv.fit_transform(new_df['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics import pairwise
similarity=pairwise.cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances=sorted(list(enumerate(similarity[index])), reverse=True , key=lambda x:x[1])
    for i in distances[1:11]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Cars 2')
```
